```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/1mg-2.csv')
df.head()
```

```python
df['key_benefits_list'][0]
```

```python
df['key_benefits_list'] = df['key_benefits_list'].str.replace("']",'')
df['key_benefits_list'] = df['key_benefits_list'].str.replace("[",'')
df['key_benefits_list'] = df['key_benefits_list'].str.replace("'",'')
```

```python
mast_ind = []
mas_ben = []
for ele in range(df.shape[0]):
  ind = []
  benefits = []
  for i in [df['key_benefits_list'][ele]]:
    w = i.split(',')
    for j in w:
      j = j.strip()
      if j.count(' ')>1:
        benefits.append(j)
      else:
        ind.append(j)
  mast_ind.append(ind)
  mas_ben.append(benefits)
 
```

```python
print(len(mast_ind))
print(len(mas_ben))
```

```python
df1 = pd.DataFrame({"Ingredients":mast_ind, 'Benefits':mas_ben})
df1 = df1.reset_index()
```
